## ResNet- Residual Networks


### Deep Residual Learning for Image Recognition

Residual model was the winner of ImageNet challenge in 2015. According to the research paper "On the ImageNet dataset we evaluate residual nets with a depth of up to 152 layers—8× deeper than VGG nets but still having lower complexity. An ensemble of these residual nets achieves 3.57% error on the ImageNet test set. This result won the 1st place on the ILSVRC 2015 classification task. We also present analysis on CIFAR-10 with 100 and 1000 layers."

Available Architecture:
    
- Resnet50 - Instantiates the ResNet50 architecture.
- Resnet101 - Instantiates the ResNet101 architecture.
- Resnet152 - Instantiates the ResNet152 architecture.
