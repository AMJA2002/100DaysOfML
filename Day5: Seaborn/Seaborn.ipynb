### Data Visualization With Seaborn

Python Seaborn library is a widely used data visualization library that is commonly used for data science and machine learning tasks. Seaborn libaray comes in handly while performing EDA. 

````python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
````

### Boxplot

A boxplot also called whisker plot is a representation that aims to provide lots of context for any given dataset. The chart visually represents several summary metrics in a single view: minimum, maximum, median, and quartile range. 

You can watch this [StatQuest](https://www.youtube.com/watch?v=fHLhBnmwUM0) to understand BoxPlot. 

````python
sns.set_style("whitegrid")
data = np.random.normal(size=(20,10)) + np.arange(10) / 2
sns.boxplot(data=data)
````

[![boxplot](https://images.squarespace-cdn.com/content/v1/55b6a6dce4b089e11621d3ed/afc2044d-4a65-4024-8d57-84a579bb92f2/test_score_box_plot.png?format=500w)](https://www.storytellingwithdata.com/blog/what-is-a-boxplot)

````python
sns.set_style("darkgrid")
data = np.random.normal(size=(20,10)) + np.arange(10) / 2
sns.boxplot(data=data)
````

### Title And Label In Seaborn

We have seen earlier in Day 4 that Title and Label are used to add the meaning to out graphical representation.

````python
sns.countplot(arr,x="Random Numbers For Count Plot",palette="ocean")
plt.title("Count Plot In Seaborn")
plt.ylabel("Count")
plt.xlabel("Numbers")
````

To increase the size of the Plot We can use `figsize`

````python
plt.figure(figsize=(10,10))
sns.set_style("whitegrid")
data = np.random.normal(size=(20,10)) + np.arange(10) / 2
sns.boxplot(data=data)
````

### Heatmap

Heatmap: A visual representation of data using colors. A heatmap can be used with all sorts of data, from representing the number of foreclosures to the spreads of credit default swaps.

Source: [Investopedia](https://www.investopedia.com/terms/h/heatmap.asp)

To check the empty data values in the data, we shall use Heatmap

````python
sns.heatmap(data.isnull())
````

This color show that there is no missing values

````python
#We also use Heatmap to check the coorelation among the data

sns.heatmap(data.corr(),cmap="gist_gray")
````

````python
sns.heatmap(data.corr()) #default
````

### Pairplot 

A pairplot graphical represenation is a pairwise relationships in a dataset. In this plot a gris of axes is created just like Correlation, every feature is compared with another feature and plotted using their relation. Here each variable in data will by shared in the y-axis across a single row and in the x-axis across a single column

````python
sns.pairplot(data,palette="bone")
````

Use Countplot to check the total count of Abnormal and Normal In the dataset

````python
sns.countplot(data=data,x="class",palette="Greens")
````

### Barplot

We already know what Barplot means from previous Day. 
It is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent

````python
x = np.array([10,20,30,40,50])
y = np.array([6,19,21,8,2])

sns.barplot(x,y)
````
