## Hotel Bookings EDA

Explore the data by analysing and visualizing

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("hotel_bookings.csv")
```

```python
data.sample(10)
```

```python
data.info()
```

```python
data.describe()
```

```python
sns.heatmap(data.isnull())
```

```python
#to check the null percentage of each row

for col in data.columns:
    null_percent = (data[col].isnull().sum()*100)/len(data)
    if null_percent>20:
        print(f"{col} has total {null_percent}% null values")
```

```python
data.isnull().sum()
```

```python
data.drop("company",axis=1,inplace=True)
```

```python
len(data)==data.shape[0]
```

```python
sns.countplot(data=data,x="hotel")
plt.xlabel("Hotel Names")
plt.title("All Hotel Names Count")
```

```python
plt.figure(figsize=(15,7))
sns.countplot(data=data,x='arrival_date_month')
plt.title("Arrival Month Count")
plt.xticks(rotation=85)
plt.xlabel("Month")
```
